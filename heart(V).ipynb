```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
sizes=[df['age'].mean(),df['trestbps'].mean(),df['chol'].mean(),df['oldpeak'].mean(),df['thalach'].mean()]
```

```python
labels=['age','trestbps','chol','oldpeak','thalach']
```

```python
explode=(0.1,0,0,0,0)
```

```python
colors=['Red','Yellow','Green','Blue','orange']
```

```python
plt.pie(sizes,labels=labels,autopct="%1.1f%%",explode=explode,colors=colors)
plt.title('Average Data')
plt.show()
```

```python
plt.hist(df['age'],bins='auto',alpha=0.7,color='blue')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Hist of age")
plt.show()
```

```python
x_line=df.iloc[:20,3]
y_line=df.iloc[:20,4]

plt.plot(x_line,marker="*",linestyle='dashed',label='Blood pressure')
plt.plot(y_line,marker="o",linestyle='dashed',label='Cholestrol')
plt.ylim(0,500)
plt.xlabel("Index")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
plt.scatter(df['age'],df['thalach'],marker='*',color='red')
plt.xlabel("Age")
plt.ylabel('Max Heart Rate')
plt.title("Age vs Max heart rate")
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.corr(),annot=True)
```

```python
sns.displot(df['age'], color='red')
plt.show()
```
